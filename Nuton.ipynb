```python
import math
from scipy.misc import derivative

def f(x):
    return 3*pow(x,4) - 4*pow(x,3) + pow(x,2) - 2*x - 3

gran_1 = 1
gran_2 = 2
eps = 0.001

def Nuton(gran_1, gran_2, eps):
    c2 = derivative(f, gran_2, n = 2)
    if (f(gran_2)*c2>0):
        x_n = gran_2
    else:
        x_n = gran_1
    d = derivative(f,x_n, n= 1)
    x_n_1 = x_n - f(x_n)/d
    while (abs(x_n_1 - x_n)>eps):
        x_n = x_n_1
        x_n_1 = x_n - f(x_n)/d
    return print ('За Ньютоном x =', x_n_1)
Nuton (gran_1, gran_2, eps)
print()
```
