```python
import math
from scipy.misc import derivative

def f(x):
    return 3*pow(x,4) - 4*pow(x,3) + pow(x,2) - 2*x - 3

def Combo(gran_1, gran_2, eps):

    if (derivative(f, gran_1, n = 1)*derivative(f, gran_1, n = 2)>0):
        a_n = gran_1
        b_n = gran_2
    else:
        a_n = gran_1
        b_n = gran_2
    a_n_i = a_n
    b_n_i = b_n

    while abs(a_n_i-b_n_i)>eps:
        a_n_i_1 = a_n_i -f(a_n_i)*(b_n_i - a_n_i)/(f(b_n_i) - f(a_n_i))
        b_n_i_1 = b_n_i - f(b_n_i)/derivative(f,b_n_i, n= 1)
        a_n_i = a_n_i_1
        b_n_i = b_n_i_1

    x = (a_n_i_1+b_n_i_1)/2

    return print('За комбінованим методом x =', x)
Combo(1, 2, 0.0001)
print()
```
