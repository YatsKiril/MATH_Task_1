```python
from random import randint

def print_array(arr):
    arr = list(map(list, zip(*arr)))
    for i in range(len(arr)):
        k = len(max(list(map(str, arr[i])),  key=len))
        arr[i] = [f'{elem:{k}d}' for elem in arr[i]]
    arr = list(map(list, zip(*arr)))
    for row in arr:
        print(*row)

n = 3
m = 3
arr = [[randint(-9, 9) for j in range(m)] for i in range(n)]
print('Матриця:')
print_array(arr)

print()
print('Справа кількість в рядку\nВнизу кількість в стовпці')
print('Справа знизу загальна кількіть')
print()

neg_col = [0]*(m+1)
for i in range(n):
    neg_row = 0
    for j in range(m):
        if arr[i][j] < 0:
            neg_row += 1
            neg_col[j] += 1
    arr[i].append(neg_row)
arr.append(neg_col)
arr[-1][-1] = sum(neg_col)

print()
print_array(arr)
```
