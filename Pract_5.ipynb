```python
import numpy as np
from scipy import optimize
from scipy.misc import derivative
import math
import matplotlib.pyplot as plt

# Числа

x0 = 0.5
y0 = -0.09
delta = 0.1

def f1(y):
    return 1 - math.cos(y) / 2
def f2 (x):
    return math.sin(x + 1) - 1
def iter (x,y,e):
    x_n = x
    y_n = y
    x_n1 = f2(x)
    y_n1 = f1(y)
    n = 1
    while ((abs(x_n1-x_n)>=e) & (abs(y_n1-y_n) >=e)):
        x_n = x_n1
        y_n = y_n1
        x_n1 = f2(y_n)
        y_n1 = f1(x_n)
        n += 1
    print ('Проста ітерація:')

    print ('x=', x_n, '\ny=',y_n,'\nКількість ітерацій = ',n)
iter(x0,y0,0.0001)

def f3(x):
    return math.sin(x[0] + 1) - x[1] - 1, 2 * x[0] + math.cos(x[1]) - 2
s = optimize.root(f3, [0.,0.], method = 'hybr')
print ('Перевірка',s.x)

# Графік

x_min, x_max = -1, 1
y_min, y_max = -4, 1
step = 0.1

x, y = np.meshgrid(np.arange(x_min, x_max, step),
                   np.arange(y_min, y_max, step))

eq1 = np.sin(x + 1) - y - 1
eq2 = 2 * x + np.cos(y) - 2

fig, a_x = plt.subplots(figsize=(10, 10))
a_x.contour(x, y, eq1, levels=[0], colors='red')
a_x.contour(x, y, eq2, levels=[0], colors='blue')

a_x.set_xlim([x_min, x_max])
a_x.set_ylim([y_min, y_max])
a_x.set_xlabel('x')
a_x.set_ylabel('y')
a_x.set_title('Графік')
plt.grid(True)

plt.show()
```
